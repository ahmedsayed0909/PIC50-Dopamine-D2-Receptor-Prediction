## **Importing libraries**

```python
import pandas as pd
from chembl_webresource_client.new_client import new_client
import matplotlib.pyplot as plt
```

## **Search for Target protein**

### **Target search for Dopamine D2 receptor**

```python
target = new_client.target
target_query = target.search('Dopamine D2 receptor')
targets = pd.DataFrame.from_dict(target_query)
targets
```

### Dataset Overview: Dopamine D2 receptor Target Classifications

In this dataset, the molecular bioactivity is categorized by the specific biological structure being targeted. Understanding these classifications is useful for determining how we handle the data for machine learning models:

* **Single Protein:** The bioactivity is measured against the standalone Dopamine D2 receptor protein or an individual related receptor (like ErbB-2 or ErbB-3). 
* **Protein-Protein Interaction:** The target is the physical interface between Dopamine D2 receptor and another distinct protein (e.g., CCN2). Compounds here are typically designed to disrupt or enhance this binding.
* **Protein Family:** The target encompasses a broader group of related receptors simultaneously, such as drugs that affect both HER1 (Dopamine D2 receptor) and HER2.
* **Protein Complex:** The target is a functional, multi-protein unit consisting of bound subunits, such as the ErbB-2/ErbB-3 heterodimer.
* **Chimeric Protein:** An engineered, hybrid protein combining domains from different sources (e.g., an MER/Dopamine D2 receptor chimera), typically used in specialized laboratory assays.

### **Select and retrieve bioactivity data for *Dopamine D2 receptor***

### we will select the first entity as it it has the perfect match
* Target type --> Single (the Dopamine D2 receptor test alone)
* organism --> Mus musculus (the test applied on humans)

```python
selected_target = targets.target_chembl_id[0]
selected_target
```

Here, we will retrieve only bioactivity data for *Dopamine D2 receptor* (CHEMBL217) that are reported as pChEMBL values.

```python
activity = new_client.activity
data = activity.filter(target_chembl_id=selected_target).filter(standard_type="IC50")
```

### we use standard type IC50 which is the standard metric used to measure a drug's potency. Specifically, it tells you exactly how much of a chemical compound is needed to inhibit a biological target (like the Dopamine D2 receptor protein) by 50%

### In a machine learning pipeline predicting molecular bioactivity for targets like the Dopamine D2 receptor protein, the IC50 value is typically your main target variable (the y value your model is trying to predict)
